## Pre-processing

# Sampling plans  

*Group 16*  
→ k1 = 3 and  
→ lots = 41, 46, 19, 56, 31, 3, 51, 32, 86, 22

```python
k1 = 3
our_lots = [41, 46, 19, 56, 31, 3, 51, 32, 86, 22]
```

*Auxiliary functions*

```python
def unique_check(lst):
    if len(lst) == len(set(lst)):
        print('We have sampled the following unique units: \n{}'.format(lst))
    else:
        print('Oops, drawn units are not unique.👎')    
```

```python
def costs(no_inspected, accept):
    k1_costs = k1 * no_inspected
    
    accept_index = [a-1 for a in accept]
    k2_costs = 0
    for i in accept_index:
        k2_costs += (200 - sum(df.iloc[i])) * 200
    
    total_costs = k1_costs + k2_costs
    return k1_costs, k2_costs, total_costs
```

### Single sampling plan

*Set-up:*

```python
our_lots_index = [a-1 for a in our_lots]
```

```python
random.seed(30) # maakt resultaten reproduceerbaar. Als we deze notebook runnen krijgen we steeds dezelfde random waarden.

accept = []
reject = []

for i in our_lots_index:
    print('For lot {}:'.format(i+1))
    lst = random.sample(range(1, 201), 45) # sample 45 (unique) values from interval [1,200] (as range(inclusive=1, exclusive=201))
    unique_check(lst)
    random_sample = [a-1 for a in lst] # -1 as unit 1 has index 0, and so on
    lot = df.iloc[i] # select lot
    lot_sample = list(lot[j] for j in random_sample)
    
    sum_lot_sample = sum(lot_sample)
    print('The sample of lot {} has {} qualitative and {} non-qualitative units.'.format(i+1, sum_lot_sample, 45-sum_lot_sample))
    
    if sum_lot_sample >= 44:
        print('We accept lot {}.✔️\n'.format(i+1))
        accept.append(i+1)
    else:
        print('We reject lot {}.❌ \n'.format(i+1))
        reject.append(i+1)

print('We accept lots {} and reject lots {}.\n'. format(accept, reject))

no_inspected = 10 * 45
k1_costs_single, k2_costs_single, total_costs_single = costs(no_inspected, accept)

print('Total costs of this single sampling plan: {}, of which {} are inspection costs and {} are caused by placing defective units in assembly.'.format(total_costs_single, k1_costs_single, k2_costs_single))
```

## Double sampling plan

```python
random.seed(123)

accept1 = []
reject1 = []
second_chance = []
samples = [] # We'll keep track of the sampled units, as we do not want to sample these again (if second chance).

for i in our_lots_index:
    print('For lot {}:'.format(i+1))
    lst = random.sample(range(1, 201), 28)
    unique_check(lst)
    random_sample = [a-1 for a in lst]
    samples.append(random_sample)
    lot = df.iloc[i]
    lot_sample = lot[random_sample]
    sum_lot_sample = sum(lot_sample)
    print('The sample of lot {} has {} qualitative and {} non-qualitative units.'.format(i+1, sum_lot_sample, 28-sum_lot_sample))
    
    if sum_lot_sample == 28:
        print('We accept lot {}.✔️\n'.format(i+1))
        accept1.append(i+1)
    elif sum_lot_sample == 27:
        print('Lot {} is questionable and gets a second chance.😐\n'.format(i+1))
        second_chance.append(i+1)
    else:
        print('We reject lot {}.❌ \n'.format(i+1))
        reject1.append(i+1)
        
print('We accept lots {}, reject lots {} and give lots {} a second chance.'. format(accept1, reject1, second_chance))

```

```python
random.seed(123)

accept2 = []
reject2 = []
second_chance2 = [a-1 for a in second_chance]
for i in second_chance2: # Twijfelachtige loten worden nog eens gesampled.
    print('For lot {}:'.format(i+1))
    used_sample = samples[our_lots_index.index(i)] # Deze units werden reeds gesampled en zullen worden verwijderd.
    total = list(range(0,200)) # O tot 199, want indices. 
    for j in used_sample:
        total.remove(j)
    new_total = total
    lst = random.sample(new_total, 26) # We samplen 26 units uit de verzameling van units die nog niet werden gesampled.
    
    unique_check(lst)
    
    lot = df.iloc[i]
    lot_sample = lot[lst]
    
    sum_lot_sample = sum(lot_sample)
    print('The second sample of lot {} has {} qualitative and {} non-qualitative units.'.format(i+1, sum_lot_sample, 26-sum_lot_sample))
    
    if sum_lot_sample == 26: ## D2 moet 0 zijn opdat lot toch nog geaccepteerd wordt.
        print('We accept lot {}.✔️\n'.format(i+1))
        accept2.append(i+1)
    else:
        print('We reject lot {}.❌ \n'.format(i+1))
        reject2.append(i+1)
        
print('We accept lots {} and reject lots {}.\n'. format(accept2, reject2))
print('As summary, in a double sampling plan, lots {} get accepted and lots {} get rejected.\n'.format(accept1+accept2, reject1+reject2))

no_inspected1 = 10 * 28 + 4*26 # 4 lots get sampled a second time
k1_costs_double, k2_costs_double, total_costs_double = costs(no_inspected1, accept1+accept2)

print('Total costs of this double sampling plan: {}, of which {} are inspection costs and {} are caused by placing defective units in assembly.'.format(total_costs_double, k1_costs_double, k2_costs_double))
```

## Continuous sampling

```python
def interval(x):
    lower = -1.095 + 0.04*x
    upper = 1.348 + 0.04*x
    return lower, upper    
```

```python
lst = random.sample(range(0,200), 200)
len(lst) == len(set(lst))
```

```python
lst[0:1]
```

```python
random.seed(123)


accept3 = []
reject3 = []
total_number_inspected = 0

for i in our_lots_index:
    print('For lot {}:'.format(i+1))
    lst = random.sample(range(0,200), 200) # We plaatsen de 200 units van elk lot in een random volgorde.
    x = 0 # cumulatief aantal geïnspecteerde units
    y = 0 # cumulatief aantal fouten
    while interval(x)[0] < y < interval(x)[1]: ## assumptie: strikte tekens
        y += (1- df.iloc[i][lst[x]]) # y + 1 als unit defect (= 0), y + 0 als unit kwalitatief
        # door lst[x] te gebruiken, gaan we steeds naar volgende, random unit in lot.
        x += 1 
        
    total_number_inspected += x
    
    unique_check(lst[0:x])
    print('We have the following (x,y) coordinates: ({},{}).'.format(x, y))
    
    if interval(x)[0] >= y:
        print('As our lower line has coordinates ({}, {}), we see that our point falls below.\nWe accept lot {}. ✔️\n'.format(x, round(interval(x)[0], 3), i+1))
        accept3.append(i+1)
    elif interval(x)[1] <= y:
        print('As our upper line has coordinates ({}, {}), we see that our point falls higher.\nWe reject lot {}. ❌\n'.format(x, round(interval(x)[1], 3), i+1))
        reject3.append(i+1)
    else: 
        print('oops')
    
print('We accept lots {} and reject lots {}.\n'. format(accept3, reject3))

k1_costs_cont, k2_costs_cont, total_costs_cont = costs(total_number_inspected, accept3)

print('Total costs of this single sampling plan: {}, of which {} are inspection costs and {} are caused by placing defective units in assembly.'.format(total_costs_cont, k1_costs_cont, k2_costs_cont))
```

# Question 3
AQL of 5% means we should only accept lots with a proportion of defectives of 5% or lower.  


```python
def evaluate(name, accept, reject):
    print('Evaluation of {} sampling plan:\n'.format(name))
    for i in accept:
        AQL = (200 - sum(df.iloc[i-1]))/200
        print('AQL of accepted lot {}: {}'.format(i, AQL))
        if AQL > 0.05:
            print('Wrongfully accepted!')
    print('\n')
    for i in reject:
        AQL = (200 - sum(df.iloc[i-1]))/200
        print('AQL of rejected lot {}: {}'.format(i, AQL))
        if AQL <= 0.05:
            print('Lot {} is wrongfully rejected!'.format(i))
    print('\n')
```

```python
evaluate('single', accept, reject)
evaluate('double', accept1+accept2, reject1+reject2) # zelfde uitkomst als single; logisch want we accepteren/weigeren dezelfde loten
evaluate('continuous', accept3, reject3)

```
